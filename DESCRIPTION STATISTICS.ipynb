```python
int1=emp.select_dtypes(include=['int64'])   
obj1=emp.select_dtypes(include=['object'])
```

```python
#linear regression
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
obj=obj1.apply(le.fit_transform)
```

```python
final=pd.concat([int1,obj],axis=1)
```

```python
final.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
l=LinearRegression()
```

```python
y=final.Attrition
x=final.drop(['Attrition'],axis=1)
```

```python
lm=l.fit(x,y)
```

```python
lm.score(x,y)
```

```python
lpredict=lm.predict(x)
```

```python
l.intercept_ 
```

```python
residual=y-lpredict
```

```python
residual
```

```python
np.sqrt(np.mean(residual**2)) #accuracy
#RMSE
```
