```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as mpl
data=pd.read_csv('D:\Studies\EmployeeAttrition.csv')
data.head()
```

```python
data.describe()
```

This command is used to calculate mean,standard deviation,minimum value,maximum value,first quatile,second quatile,third quatile and maximum value of every column in the given datafile. 

```python
data.groupby('Attrition').mean()
```

This command is used to calculate the mean value based on the 'attrition' attribute values.From the output data we can infer that most of the employees didn't satisfy with their job because the attrition mean of 'No' value is greater than mean of 'yes' value under Environment_Satisfication,Job_involvement,Relationship_satisfication.

```python
data.groupby(['Department','Attrition']).size()
```

From the above output,We can infer that in every department the attrition value of 'no' is greater than the attrition value of 'yes' and also we can also see that there are more employees in the Reasearch and Development department

```python
pd.crosstab(data.Gender,data.Attrition).plot(kind='bar')
mpl.title('Attrition based on Gender attribute')
mpl.xlabel('Gender')
mpl.ylabel('Frenquency of Attrition')
```

This command is used to view the bar chart for the attrition based on the gender attribute.Based on that more male employees have been attritioned than the female employees.

```python
pd.crosstab(data['EducationField'],data['Attrition']).plot(kind='bar')
mpl.title('Attrition with respect to EducationField')
mpl.xlabel('EducationField')
mpl.ylabel('Frequency of Attrition')
```

Based on the above bar chart,Life science and medical fields have highest number of employees and also attrition rates.

```python
pd.crosstab(data['MaritalStatus'],data['Attrition']).plot(kind='bar')
mpl.title('Attrition with respect to MaritalStatus')
mpl.xlabel('MaritalStatus')
mpl.ylabel('Frequency of Attrition')
```

Based on the above bar graph,among married and divorced people there is lower rate of attrition when compared to employees who are single.
