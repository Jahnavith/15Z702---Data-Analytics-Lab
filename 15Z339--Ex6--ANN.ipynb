```python
import pandas as pd
import numpy as np
df = pd.read_csv("movie_metadata.csv")
df.head()
```

```python
x=df[['imdb_score','director_facebook_likes','actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes','movie_facebook_likes','num_user_for_reviews','num_critic_for_reviews']]
x.head()
```

```python
df.isnull().sum()
```

```python
x = x.dropna()
x = x [x.actor_2_facebook_likes!=0.0]
x = x [x.actor_3_facebook_likes!=0.0]
x = x [x.movie_facebook_likes!=0.0]
x = x [x.director_facebook_likes!=0.0]
x = x [x.num_critic_for_reviews!=0.0]
x = x [x.actor_1_facebook_likes!=0.0]
x = x [x.num_user_for_reviews!=0.0]
x = x [x.imdb_score!=0.0]
df1 = x[['num_critic_for_reviews','director_facebook_likes','actor_1_facebook_likes','movie_facebook_likes','num_user_for_reviews']]
y = x[['imdb_score']]
y = y.round()
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(df1, y, test_size = 0.15)
```

```python
from sklearn.neural_network import MLPClassifier  
mlp = MLPClassifier(hidden_layer_sizes=(6,4,2), max_iter=1000,random_state=1,)  
mlp.fit(X_train, y_train.values.ravel())
```

```python
predictions = mlp.predict(X_test) 
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
