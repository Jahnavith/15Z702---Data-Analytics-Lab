```python
import numpy as np
import pandas as pd
from sklearn import preprocessing

#import library for agglomerative clustering
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt

data = pd.read_csv("iris.csv",sep=',',header=0)

#converting the categorical attribute to continous/numerical attribute
if data.species.dtype == 'object':
    lbl = preprocessing.LabelEncoder()
    lbl.fit(list(data.species.values))
    data.species = lbl.transform(list(data.species.values))


#clustering using single linkage
single_Z = hierarchy.linkage(data, 'single')
plt.figure()
dn = hierarchy.dendrogram(single_Z)
```

```python
#clustering using complete linkage
complete_Z = hierarchy.linkage(data,'complete')
plt.figure()
dn = hierarchy.dendrogram(complete_Z)
```

```python
#clustering using average linkage
average_Z = hierarchy.linkage(data,'average')
plt.figure()
dn = hierarchy.dendrogram(average_Z)
```
