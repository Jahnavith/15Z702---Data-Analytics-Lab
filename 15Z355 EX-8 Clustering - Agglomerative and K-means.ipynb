```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
X = iris.data
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
clt = AgglomerativeClustering(linkage='complete', 
                              affinity='euclidean', 
                              n_clusters=3)

model = clt.fit(X_std)
```

```python
model.labels_
```

```python
linkage_matrix = linkage(X, 'ward')
figure = plt.figure(figsize=(7.5, 5))
dendrogram(
    linkage_matrix,
    truncate_mode='lastp',  
    p=150,                  
    leaf_rotation=90.,      
    leaf_font_size=8.,
)
plt.title('Hierarchical Clustering Dendrogram - Single(Ward)')
plt.xlabel('sample index')
plt.ylabel('distance')
plt.tight_layout()
plt.show()
```

```python
linkage_matrix = linkage(X, 'complete')
figure = plt.figure(figsize=(7.5, 5))
dendrogram(
    linkage_matrix,
    truncate_mode='lastp',  
    p=150,                  
    leaf_rotation=90.,      
    leaf_font_size=8., 
)
plt.title('Hierarchical Clustering Dendrogram - Complete')
plt.xlabel('sample index')
plt.ylabel('distance')
plt.tight_layout()
plt.show()
```

```python
linkage_matrix = linkage(X, 'average')
figure = plt.figure(figsize=(7.5, 5))
dendrogram(
    linkage_matrix,
    truncate_mode='lastp',  
    p=150,                  
    leaf_rotation=90.,      
    leaf_font_size=8.,
)
plt.title('Hierarchical Clustering Dendrogram - Average')
plt.xlabel('sample index')
plt.ylabel('distance')
plt.tight_layout()
plt.show()
```
