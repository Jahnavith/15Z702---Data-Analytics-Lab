### Using k-means

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

iris = load_iris()
X = iris.data

kmeans = KMeans(n_clusters=4, random_state=1).fit(X)
kmeans.labels_
```

### 2) Using Agglomerative Clustering


```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
import numpy as np

def plot_dendrogram(model, **kwargs):
    children = model.children_
    distance = np.arange(children.shape[0])
    no_of_observations = np.arange(4, children.shape[0]+4)
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)
    dendrogram(linkage_matrix, **kwargs)
```

#### a) Single Linkage

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(linkage="ward",n_clusters=7)
singleLinkage = model.fit(X)

plt.figure(figsize=(16,8))
plt.title('Single Linkage')
plot_dendrogram(model, labels=singleLinkage.labels_)
plt.show()
```

#### b) Complete linkage

```python
model = AgglomerativeClustering(linkage="complete",n_clusters=7)
completeLinkage = model.fit(X)

plt.figure(figsize=(16,8))
plt.title('Complete Linkage')
plot_dendrogram(model, labels=completeLinkage.labels_)
plt.show()
```

#### c) Average linkage

```python
model = AgglomerativeClustering(linkage="average",n_clusters=7)
averageLinkage = model.fit(X)

plt.figure(figsize=(16,8))
plt.title('Average Linkage')
plot_dendrogram(model, labels=averageLinkage.labels_)
plt.show()
```
