## Perform web scrapping on PSG IM website to display names of faculty with Ph.D


Importing the required Packages

```python
import requests
from bs4 import BeautifulSoup
import re
```

Collecting the required Webpage 

```python
page = requests.get('http://psgim.ac.in/2017/01/full-time-faculty/')
```

Parse the webpage

```python
soup = BeautifulSoup(page.text,'html.parser')
```

Print the parse tree

```python
print(soup.prettify())
```

Finding all the 'a' tags

```python
faculty = soup.find_all('a')
```

Declaring a regular expression

```python
Qual = "^Dr."
```

Using the get_text(), we can get all the contents and we have to match with required pattern and print them

```python
for fac in range(len(faculty)):
    a = faculty[fac].get_text()
    if(re.match(Qual,a)):
        print(a)
    
```
